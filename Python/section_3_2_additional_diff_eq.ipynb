````python
x1 = 1
n = 10
a = 2
x = np.repeat(0, n)
x[0] = x1

for i in np.arange(1, n):
    x[i] = x[i-1] * a

print(x)

plt.plot(x, marker="o")
plt.show()
````

````python
n = 30
x1 = 1
x2 = 2
mySd = .008
myNoise = 100 * np.random.normal(scale=mySd, size=n)
x = [x1, x2] + myNoise[:2]
a1 = 1
a2 = -.8

for i in np.arange(2, n):
    deterministicRecursivePart = a1 * x[i-1] + a2 * x[i-2] 
    innovation = myNoise[i]
    newValue = deterministicRecursivePart + innovation
    x = np.append(x, newValue)

print(x)

plt.plot(x, marker="o")
plt.show()
````
