````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf 
````

In R:
```
# set.seed(1) # set seed to reproduce results below
x <- rnorm(5)
y <- 1*x + rnorm(5)
fit <- lm(y ~ x) # fit linear model
# we can get summary for our model (residuals, coefficients, R-squared, etc.)
summary(fit)
```

````python
np.random.seed(1)
x = np.random.normal(size=5)
y = 1*x + np.random.normal(size=5)
df = pd.DataFrame({"x": x, "y": y})
fit = smf.ols("y~x", data=df).fit()
print(fit.summary())
````

In R:
```
we can extract necessary information from lm object using one of two forms below
fit$residuals # get residuals from fitted model
resid(fit)
```

````python
fit.resid
````

In R:
```
fit$coefficients # get model coefficients
coef(fit)
```

````python
fit.params
````

In R:
```
fit$fitted.values # get fitted values from the model
fitted(fit)
```

````python
fit.fittedvalues
````

In R:
```
# plot our points and then with abline function add fitted line
plot(x, y)
abline(fit)
```

````python
fig = sm.graphics.abline_plot(model_results=fit, color="red")
ax = fig.axes[0]
ax.scatter(x, y)
plt.show()
````
